```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

$Найдем\quad  производную\quad  вперед\quad  по\quad  формуле\quad  f'(x)=\frac{f(x+h)-f(x)}{h}$ 

$
h=2*\sqrt{\frac{M_0*E_м}{M_2}},\quad  где\quad  в\quad  нашем\quad  случае\quad  M_0=M_2 - max\quad функции
$

```python
x = np.linspace(0,math.pi*2,100)
M = np.sin(x).max()
h = 2*math.sqrt((M*2**(-53))/M)

y = (np.sin(x+h)-np.sin(x))/h
plt.plot(x,y,'r')
plt.title('d(sin)')
plt.show()
```

Сравним график производной с графиком $cos(x)$

```python
h = 2*math.sqrt(2**(-53))
x = np.linspace(0,math.pi*2,100)
y = (np.sin(x+h)-np.sin(x))/h
plt.plot(x,y,'r*')

x = np.linspace(0,math.pi*2,100)
y = np.cos(x)
plt.plot(x,y,'g')

plt.show()
```

Как видно из сравнения графиков, производная была найдена точно.
