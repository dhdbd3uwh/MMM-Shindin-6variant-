## Задание 1
#### Метод дихотомии 

```python
x1=-2
x2=0
E=10**-4
n=0

def f(x):
    return x**4-18*(x**3)-10

while abs(x2-x1)>E:
    c=(x1+x2)/2
    if(f(x1)*f(c)<0):
        x2=c
    else:
        x1=c
    n+=1
    
print("с=",c)
print("f(c)=",f(c))
print("n=",n)
```

#### Метод хорд

```python
x1=-2
x2=0
n=0

c=x1-(x2-x1)/(f(x2)-f(x1))*f(x1)
while abs(f(c))>E:
    c=x1-(x2-x1)/(f(x2)-f(x1))*f(x1)
    if(f(x1)*f(c)<0):
        x2=c
    else:
        x1=c
    n+=1
    
print("c=",c)
print("f(c)=",f(c))
print("n=",n)
```
